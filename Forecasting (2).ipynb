```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt


df1 = pd.read_csv("CocaCola_Sales_Rawdata.csv")
df1.head()
```

```python
df1.isnull().sum()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df1['Sales'])
plt.title('CocaCola Prices Over Time')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.show()
```

```python
model = ARIMA(df1['Sales'])
model_fit = model.fit()
```

```python
forecast = model_fit.get_forecast(steps=len(df1))
```

```python
rmse = sqrt(mean_squared_error(df1['Sales'], forecast.predicted_mean))
print(f'RMSE for CocaCola Prices Forecasting: {rmse}')
```

```python
df2 = pd.read_csv("Airlines+data.csv")
df2.head()
```

```python
df2.isnull().sum()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df2['Passengers'])
plt.title('Airlines Passengers Over Time')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model = SARIMAX(df2['Passengers'])
model_fit = model.fit()
```

```python
forecast = model_fit.get_forecast(steps=len(df2))
```

```python
rmse = sqrt(mean_squared_error(df2['Passengers'], forecast.predicted_mean))
print(f'RMSE for Airlines Passengers Forecasting: {rmse}')
```

### we can choose the 2nd model with the lower RMSE for forecasting.
